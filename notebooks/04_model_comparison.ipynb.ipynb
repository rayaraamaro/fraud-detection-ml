## Parte 4 — Comparação Crítica entre os Modelos

### Desempenho (métricas)

Comparando os dois modelos, a regressão logística apresentou desempenho superior na detecção de fraudes.

O modelo supervisionado conseguiu identificar a maior parte das transações fraudulentas, especialmente após o ajuste com class_weight='balanced', reduzindo significativamente o número de falsos negativos.

Por outro lado, o Isolation Forest apresentou maior dificuldade em detectar fraudes, com um número elevado de falsos negativos, principalmente em sua configuração inicial.

Mesmo após ajustes de parâmetros, como o aumento do contamination, o modelo não supervisionado não alcançou o mesmo nível de desempenho da regressão logística.

### Robustez ao desbalanceamento

A regressão logística demonstrou maior robustez ao desbalanceamento dos dados quando ajustada com o parâmetro class_weight='balanced'. Esse ajuste permitiu que o modelo desse maior importância à classe minoritária, melhorando a detecção de fraudes.

Já o Isolation Forest não possui um mecanismo direto para lidar com desbalanceamento, dependendo do parâmetro contamination para estimar a proporção de anomalias. No entanto, essa abordagem é menos precisa, pois assume uma taxa fixa de fraude, o que nem sempre corresponde à realidade.

Dessa forma, a regressão logística se mostrou mais adaptável ao problema de classes desbalanceadas.

### Interpretabilidade

A regressão logística é um modelo altamente interpretável, permitindo analisar o impacto de cada variável na decisão final por meio de seus coeficientes.

Por outro lado, o Isolation Forest apresenta baixa interpretabilidade, uma vez que sua decisão é baseada em múltiplas árvores e processos de isolamento, dificultando a compreensão do motivo pelo qual uma transação foi classificada como anômala.

Assim, a regressão logística oferece maior transparência, o que é importante em aplicações financeiras.

### Dependência de rótulos

A regressão logística depende diretamente de dados rotulados para aprender a distinguir entre transações normais e fraudulentas.

Em contraste, o Isolation Forest não utiliza rótulos durante o treinamento, sendo capaz de identificar padrões anômalos sem conhecimento prévio das classes.

Essa característica torna o modelo não supervisionado útil em cenários onde os dados não estão rotulados ou onde novas fraudes ainda não foram identificadas.

### Aplicabilidade no mundo real

Em cenários reais, a regressão logística tende a ser mais eficaz quando há dados históricos rotulados disponíveis, pois consegue aprender padrões específicos de fraude.

Por outro lado, o Isolation Forest pode ser útil como complemento, atuando na detecção de comportamentos anômalos que ainda não foram classificados como fraude.

Assim, uma abordagem híbrida, combinando modelos supervisionados e não supervisionados, pode ser mais eficaz para sistemas de detecção de fraude.

### Comparação das matrizes de confusão

Ao comparar as matrizes de confusão dos dois modelos, observa-se uma diferença clara nos padrões de erro.

A regressão logística, especialmente com balanceamento, apresentou menor número de falsos negativos, indicando maior capacidade de detectar fraudes. No entanto, isso ocorreu com um aumento nos falsos positivos, evidenciando o trade-off entre precisão e recall.

Já o Isolation Forest apresentou comportamento mais conservador em sua configuração inicial, com poucos falsos positivos, mas muitos falsos negativos, o que é inadequado para o problema.

Ao aumentar o parâmetro contamination, o modelo passou a detectar mais fraudes, reduzindo falsos negativos, mas ao custo de um aumento significativo nos falsos positivos.

Isso demonstra que ambos os modelos apresentam trade-offs, porém a regressão logística permite um controle mais eficaz desse equilíbrio.
